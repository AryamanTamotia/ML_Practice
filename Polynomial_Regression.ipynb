```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Importing the data
## x = data; y = quadratic Equation

x = np.array( 7 * np.random.rand(100,1) - 3)
x1 = x.reshape(-1,1)
y = 13 * x*x + 2*x + 7
```

```python
# Data Points

plt.scatter(x,y, s = 10)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Non-Linear Data')
```

```python
# Model Intialisation
regression_model = LinearRegression()
# Fit the data ( train the model )
regression_model.fit(x1,y)
print("Slope of line is ", regression_model.coef_)
print("Intercept value is ", regression_model.intercept_)

# Predicted
y_predicted = regression_model.predict(x1)
```

```python
# Data Points
plt.scatter(x,y, s = 10)
plt.xlabel("$x$", fontsize = 18)
plt.ylabel("$y$",rotation = 90, fontsize = 18)
plt.title("Data points")
# Predicted values
plt.plot(x,y_predicted, color = 'g')
```

```python
mse = mean_squared_error(y,y_predicted)
rmse = np.sqrt(mse)
r2 = r2_score(y,y_predicted)

# printing values
print("MSE of linear model ", mse)
print("R2 score of linear model ", r2)
```

```python
########################################

# Poly Regression model

poly_features = PolynomialFeatures(degree = 2, include_bias = False)
x_poly = poly_features.fit_transform(x1) 
x[3]
x_poly[3]
```

```python
lin_reg = LinearRegression()
lin_reg.fit(x_poly,y)
print("Coefficients of x are ", lin_reg.coef_)
print("Intercept is ", lin_reg.intercept_)
print("Score (Prediction) is ",lin_reg.score(x_poly,y))

# score*100 = accuracy
```

```python
x_new = np.linspace(-3,4,100).reshape(-1,1)
x_new_poly = poly_features.transform(x_new)
y_new = lin_reg.predict(x_new_poly)

plt.plot(x,y,"b.")

plt.plot(x_new, y_new, "r-", linewidth = 2,label = "Predictions" )
plt.xlabel("$x_1$",fontsize=18)
plt.ylabel("$y$",rotation=0,fontsize=18)
plt.legend(loc="upper left",fontsize = 18)
plt.title("Quadratic prediction plot")
plt.show()

y_deg2 = lin_reg.predict(x_poly)
# model evaluation
mse_deg2 = mean_squared_error(y,y_deg2)
r2_deg2 = r2_score(y_deg2,y)

# printing values
print("MSE of PolyRegression Model ", mse_deg2 )
print("R2 score of linear model ",r2_deg2 )
```
