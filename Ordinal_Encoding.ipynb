```python
import pandas as pd
import datetime
```

```python
# create a variable with dates and from that extract thw weekday
# creatw a list of dates with 20days difference from today
# and then transform it into a dataframe

df_base = datetime.datetime.today()
df_date_list = [df_base - datetime.timedelta(days=x) for x in range (0,20)]
df = pd.DataFrame(df_date_list)
df.columns = ['day']
df
```

```python
# Extract the work day name

df['day_of_work'] = df['day'].dt.weekday_name
df.head()
```
